```python
# converting labels in Numeric form
# used in classification problems
# Here we are going to work with 2 diff datasets
# iris dataset tells that the given iris flower belongs to which species
# breast cancer tells that the given breast cancer is malignant or benign
```

Label encoding of Breast cancer dataset

```python
# loading the data from csv file to pandas
cancer_data = pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
# finding datapoints/count of diff labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
# 0--> benigm
# 1--> malignant

cancer_data['target'].value_counts()
```

Label encoding for Iris dataset

```python
iris_dataset = pd.read_csv('/content/iris_data.csv')
iris_dataset.head()
```

```python
iris_dataset['Species'].value_counts()
```

```python
label_encoder_iris = LabelEncoder()
```

```python
labels_iris = label_encoder_iris.fit_transform(iris_dataset.Species)
```

```python
iris_dataset['target'] = labels_iris
```

```python
iris_dataset['target'].value_counts()
```
